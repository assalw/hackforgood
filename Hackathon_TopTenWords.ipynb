```python
# removing stop words
from stop_words import get_stop_words
stop_words = get_stop_words('english')
lenWord = 2

def removeStopWords(stop_words, txt):
    newtxt = ' '.join([word for word in txt.split() if word not in stop_words and len(word)>lenWord])
    return newtxt
```

```python
text = dh.iloc[8,4]
filename = dh.iloc[8,0]
print(filename)
print(text)
```

```python
stopText = removeStopWords(stop_words,text)
#print (stopText)
```

```python
stopText = str.lower(stopText)
#stopText
```

```python

stopText = re.sub("[^a-z]", ' ', stopText ) # remove a to z again - result from the plot
word_list = nltk.word_tokenize(stopText)
# Applying Set

ordered_tokens = set()
result = []
for word in word_list:
    if word not in ordered_tokens:
        ordered_tokens.add(word)
        result.append(word)
     
#print(result)
```

```python
def map_book(tokens):
    hash_map = {}

    if tokens is not None:
        for element in tokens:
            # Remove Punctuation
            word = element.replace(",","")
            word = word.replace(".","")

            # Word Exist?
            if word in hash_map:
                hash_map[word] = hash_map[word] + 1
            else:
                hash_map[word] = 1

        return hash_map
    else:
        return None
```

```python
# Create a Hash Map (Dictionary)
map = map_book(word_list)

#word_freq = pd.DataFrame(columns = ["word", "frequency"])
#word_freq
# Show Word Information
#for word in word_list:
       
 #   print('[' + word + '] Frequency: ' + str(map[word]))
```

```python
#next(iter(map.values()))
#map.values()
#map.keys()

list_keywords = []
list_counts = []
for i in map.values():
    list_keywords.append(i)
    
for i in map.keys():
    list_counts.append(i)
    

```

```python
dfText = pd.DataFrame.from_dict(map, orient='index', columns = ["count"])
dfText.sort_values(by='count', ascending=False, inplace = True)
dfText_top100 = dfText.nlargest(100, "count")
print (dfText_top100)
```

```python
word_freq = pd.DataFrame(columns = ["filename","list_words", "list_frequency"])
word_freq["filename"] = "xx"
word_freq["list_words"] = list_keywords
word_freq["list_frequency"] = list_counts

word_freq = pd.concat([word_freq, word_freq])
```
